```python
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob

%matplotlib inline
```

```python
paths = glob('../logs/*.csv')
print('Found {} models'.format(len(paths)))
```

```python
logs = []
for path in paths:
    log = pd.read_csv(path)
    plt.plot(log.valid_ious)
    logs.append(log)
    print(path, log['valid_ious'].max())
```
